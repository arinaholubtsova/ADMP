Task 7

```python
yes = df[df.y == 'yes']
no = df[df.y == 'no']

yes_d = yes.job.value_counts()
no_d = no.job.value_counts()
print(" \tYes:\n", yes_d, "\n\n\n", "\tNo:\n", no_d)

print("\nThe probability of signing a deposit by a person from the admin group is ", yes_d[0] / (yes_d[0] + no_d[0]) * 100, "%")

```

Task 8

```python
yes_d, no_d = yes.job.value_counts(), no.job.value_counts()
print("\tjob:")
(yes_d / (yes_d + no_d) * 100).sort_values(ascending = False)
```

Task 9

```python
yes_d2, no_d2 = yes.education.value_counts(), no.education.value_counts()
print("\teducation:")
(yes_d2 / (yes_d2 + no_d2) * 100).sort_values(ascending = False)
```
